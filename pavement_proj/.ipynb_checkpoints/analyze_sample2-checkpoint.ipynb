```python
import pandas as pd

SIGNIFICANCE_LEVEL = 0.05
EQUIVALENCE_INTERVAL = 5

data_df = pd.read_excel(r'./CAPRI Data.xlsx')

data_df['Condition_Diff'] = data_df['Condition A'] - data_df['Condition B']
data_df['HMA_WMA'] = data_df['HMA/WMA'].astype('category')
data_df['RAP'] = data_df['RAP'].astype('category')
data_df['NMAS'] = data_df['NMAS'].astype('category')

data_df
```

```python
# Equality-line assessment
import plotly.express as px
import numpy as np
import plotly.graph_objects as go
from scipy import stats

def equality_line_assessment(subset: pd.Series, name: str, show_legend = False) -> tuple[go.Figure, bool, float, float]:
    df = data_df[subset]
    x = df['Condition A']
    y = df['Condition B']
    n = len(x)
    lin_reg_result = stats.linregress(x, y)

    t_crit = stats.t.ppf(1-SIGNIFICANCE_LEVEL/2, df=n-2)
    ci_low = lin_reg_result.slope - t_crit * lin_reg_result.stderr
    ci_high = lin_reg_result.slope + t_crit * lin_reg_result.stderr
    within_bounds = (ci_low >= 0.95) and (ci_high <= 1.05)

    equality_line = px.scatter(df, x='Condition A', y='Condition B', 
                    hover_data=['RAP', 'NMAS', 'HMA_WMA'],
                    title=f'Equality Plot for {name} subset: Condition A vs Condition B')

    min_val = min(df[['Condition A', 'Condition B']].min())
    max_val = max(df[['Condition A', 'Condition B']].max())

    line_1to1 = np.linspace(min_val, max_val, 10)
    upp_bound = 1+SIGNIFICANCE_LEVEL
    low_bound = 1-SIGNIFICANCE_LEVEL
    equality_line.add_trace(go.Scatter(x=line_1to1, y=line_1to1*(upp_bound),
                            mode='lines', name=f'{upp_bound} upper bound', showlegend=show_legend, 
                            line=dict(dash='dash', color='red')))
    equality_line.add_trace(go.Scatter(x=line_1to1, y=line_1to1*(low_bound),
                            mode='lines', name=f'{low_bound} lower bound', showlegend=show_legend, 
                            line=dict(dash='dash', color='red')))


    y_line = lin_reg_result.slope * line_1to1 + lin_reg_result.intercept 
    equality_line.add_trace(go.Scatter(x=line_1to1, y=y_line,
                            mode='lines', name=f'Linear Regression', showlegend=show_legend, 
                            line=dict(color='blue')))

    return equality_line, [name, ci_low, ci_high, within_bounds]
```

```python
hma_equality_plot, hma_ci = equality_line_assessment(data_df['HMA_WMA'] == 'HMA', 'HMA', show_legend=True)
wma_equality_plot, wma_ci = equality_line_assessment(data_df['HMA_WMA'] == 'WMA', 'WMA')
rap0_equality_plot, rap0_ci = equality_line_assessment(data_df['RAP'] == 0, 'RAP0')
rap15_equality_plot, rap15_ci = equality_line_assessment(data_df['RAP'] == 15, 'RAP15')
nmas12_equality_plot, nmas12_ci = equality_line_assessment(data_df['NMAS'] == 12.5, 'NMAS12.5')
nmas19_equality_plot, nmas19_ci = equality_line_assessment(data_df['NMAS'] == 19, 'NMAS19')
equality_data = [hma_ci, wma_ci, rap0_ci, rap15_ci, nmas12_ci, nmas19_ci]

pd.DataFrame(equality_data, columns=['factor', 'CI low bound', 'CI upper bound', 'is within [0.95, 1.05]'])
```

```python
from plotly.subplots import make_subplots

fig = make_subplots(rows=3, cols=2, 
                    subplot_titles=['HMA', 'WMA', 'RAP0', 'RAP15', 'NMAS12.5', 'NMAS19'])

fig.update_layout(title_text='Equality Line Assessment') 

[fig.add_trace(t, row=1, col=1) for t in hma_equality_plot.data]
[fig.add_trace(t, row=1, col=2) for t in wma_equality_plot.data]
[fig.add_trace(t, row=2, col=1) for t in rap0_equality_plot.data]
[fig.add_trace(t, row=2, col=2) for t in rap15_equality_plot.data]
[fig.add_trace(t, row=3, col=1) for t in nmas12_equality_plot.data]
[fig.add_trace(t, row=3, col=2) for t in nmas19_equality_plot.data]
fig.show()
```

```python
from statsmodels.stats.weightstats import ttost_paired

tost_p, _, _ = ttost_paired(data_df['Condition A'], data_df['Condition B'], -EQUIVALENCE_INTERVAL, EQUIVALENCE_INTERVAL)

print(f"Overall TOST p-value = {tost_p}")
if tost_p < SIGNIFICANCE_LEVEL:
    print(f"The two conditions are equivalent within ±{EQUIVALENCE_INTERVAL}.")
else:
    print(f"Not enough evidence to claim equivalence within ±{EQUIVALENCE_INTERVAL}.")

```

```python
# One way ANOVA
import statsmodels.api as sm
from statsmodels.formula.api import ols

def anova(factor: str):
    model = ols(f'Condition_Diff ~ {factor}', data=data_df).fit()
    anova_table = sm.stats.anova_lm(model, typ=2)
    return anova_table

[
    anova('C(RAP)'),
    anova('C(NMAS)'),
    anova('C(HMA_WMA)'),
]
```

```python
# Factorial ANOVA
anova_factorial = anova('C(RAP) * C(NMAS) * C(HMA_WMA)')

anova_factorial
```

```python
ancova = anova('C(HMA_WMA) + RAP + NMAS')

ancova
```

```python
import itertools

class Factor:
    is_category = False
    formula: str
    column: str
    def __init__(self, form_str: str):
        self.formula = form_str
        if form_str.startswith('C(') and form_str.endswith(')'):
            self.is_category = True
            self.column = form_str.removeprefix('C(').removesuffix(')')
        else:
            self.column = form_str

    

def two_way_interaction(interaction: list[Factor], controls: list[Factor], ignore_p=False):
    factors = interaction + controls
    if len(interaction) != 2:
        raise ValueError('Can only eval two way interaction')
    if not all([i.column in data_df.columns for i in factors]):
        raise ValueError(f'Interaction and controls {[i.column for i in factors]} must be in dataframe columns: {data_df.columns}')
    interaction_str = f'{interaction[0].formula} * {interaction[1].formula}'
    factor_str = ' + '.join([interaction_str, *[c.formula for c in controls]])
    model_str = f'Condition_Diff ~ {factor_str}'
    print(f'modeling: {model_str}')
    model_int = ols(model_str, data=data_df).fit()
    int_anova = sm.stats.anova_lm(model_int, typ=2)
    print(f"\nInteraction analysis ({interaction[0].column} × {interaction[1].column}):\n", int_anova)
    p_interaction = int_anova.loc[f'{interaction[0].formula}:{interaction[1].formula}', 'PR(>F)']

    if p_interaction < SIGNIFICANCE_LEVEL or ignore_p:
        print(f"Significant interaction detected between {interaction[0].column}:{interaction[1].column}.")
        for mix in data_df[interaction[1].column].unique():
            subset = data_df[data_df[interaction[1].column] == mix]
            if len(subset) > 2: 
                model_simple = ols(f'Condition_Diff ~ C({interaction[0].column})', data=subset).fit()
                print(f"\nSimple effect of {interaction[0].column} within {mix}:")
                print(model_simple.summary().tables[1])
        
        def create_df_with_interaction_permutations_and_constant_control_val(df: pd.DataFrame, interactions:list[Factor], controls: list[Factor]):
            grid_data = list(itertools.product(
                df[interactions[0].column].unique().tolist(), 
                df[interactions[1].column].unique().tolist()
            ))
            prediction_df = pd.DataFrame(grid_data, columns=[interactions[0].column, interactions[1].column])
            for c in controls:
                control_value = data_df[c.column].mode()[0] if c.is_category else data_df[c.column].mean()
                prediction_df[c.column] = control_value
            return prediction_df
        prediction_df = create_df_with_interaction_permutations_and_constant_control_val(data_df, interaction, controls)
        prediction_df['pred'] = model_int.predict(prediction_df)
        controls_str = ', '.join([c.column for c in controls])
        fig = px.line(prediction_df, x=interaction[0].column, y='pred', color=interaction[1].column, markers=True,
                        title=f'Interaction Plot: {interaction[0].column} × {interaction[1].column} (adjusted for {controls_str})')
        fig.show()
    else:
        print(f"No significant interaction between {interaction[0].column}:{interaction[1].column}")
```

```python
two_way_interaction([Factor('C(RAP)'), Factor('C(HMA_WMA)')], [Factor('C(NMAS)')], True)
```
