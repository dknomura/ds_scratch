# Pavement Factor Analysis
In this analysis, RAP is treated as a category with RAP0 and RAP15, as well as NMAS with NMAS12.5 and NMAS 19. Since HMA and WMA are naturally categorical, there are no continuous factors in this analysis. A significance level of 0.05 was used throughout the analysis. 

Null Hypothesis (H_0^I ): The measured properties of asphalt mixtures in condition A are not statistically different from those of mixtures in condition B.

Alternative Hypothesis (H_A^I): The measured properties of asphalt mixtures in condition A are statistically different from those of mixtures in condition B.  


## Equality Line Assessement
The data is broken into subsets for each factor category, i.e. RAP0, RAP15, NMAS12.5, NMAS19, HMA, and WMA, and calculated the linear regression of Condition A vs Condition B for each subset. With the slope and stderr, along with tcrit, we can find the confidence interval and see that none of the subsets' confidence interval are entirely contained within [0.95, 1.05], which is visually confirmed when plotting the linear regression with the +-5% equality line. All factors are flagged to analyze their influence on the difference between the two conditions.

## Two one sided test (TOST)
A margin of +/-5 was chosen for the TOST, but further discussion is required to determine a more appropriate equivalence interval. The p-value of the TOST was 0.278, which is greater than our 0.05 significance level, so we cannot claim that condition A and condition B are equivalent within the +/-5 margin.

## ANOVA
### One way ANOVA
To isolate the effect of each factor, a one way ANOVA test was conducted to see the effect that each factor had on the difference between condition A and condition B. All of the p-values were greater than 0.05, so there is no statistical significance that any factor has on the difference between the two conditions. 

### Factorial ANOVA
A three way ANOVA test was conducted to determine the interaction between all of the factors. However there was not enough data for this test, so the results are inconclusive. 

## ANCOVA
As noted in the introduction, all of the factors are treated as categories, therefore an ANCOVA test cannot be conducted. In theory, RAP and NMAS could be treated as continuous factors, but the two values are not enough to establish a linear relationship. 

If there was a continuous factor, then ANCOVA would adjust for the variance caused by the continuous factors and give a more accurate p-value for the categorical factor's effect. It would also reveal any linear relationship between the continuous factors and the variance between the two conditions.

## Interaction Analysis
With three different factors, there are 3 different combinations (NMAS:RAP, RAP:HMA/WMA, and NMAS:HMA/WMA). Each of these interactions along with the third factor was modeled in a two way ANOVA test. All of these tests had a P-value greater than .05, which shows that there is no statistically significant interaction between any two of the factors that impacts the difference between condition a and condition b. 

Just to show further analysis, I designed the code to ignore the >.05 P-values and continue with analysis. It splits the data into subsets based on one of the interaction factors categories and uses the other interaction factor for a one way ANOVA test. The RAP*HMA/WMA subset analysis cannot be evaluated because HMA only contains RAP0 and WMA only contains RAP15. All of the other subset analysis tests has P values > 0.05, which is consistent with the interaction analysis. I also plotted the interaction, which is only for demo purposes.  

## Tukey HSD
 There are no significant findings in any of the ANOVA tests, so there is no need for any post hoc tests. Also, since all of the factors had only 2 categories, RAP0/RAP15, NMAS12.5/NMAS19, and HMA/WMA, the Tukey HSD test is not necessary. However, if an ANOVA test was significant and the categorical factors had more than 2 categories, then we could use the Tukey HSD to see a pairwise comparison of all of the categories. If there is a statistical significance between two category pairs and the confidence interval does not contain 0, then we can conclude that the two categories differ significantly. 

## Conclusion
Since the data points were randomly generated and the sample size was small, either the tests were inconclusive or no significant differences were detected between conditions A and B in my analysis. 

The code was designed to easily accomodate more factors, but more work will need to be done to work with continuous factors and more response variables.
