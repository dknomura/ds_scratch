```python
import pandas as pd
import numpy as np

AVE_SPEED = 60
stdev = 5
NUM_TRIALS= 30
mean_speeds = {"num_vehicles": np.array([]), "tms": np.array([]), "sms": np.array([])}

for idx, x in enumerate(range(NUM_TRIALS)):
    num_vehicles = idx + 1
    speeds = np.random.normal(AVE_SPEED, stdev, num_vehicles)
    mean_speeds["num_vehicles"] = np.append(mean_speeds["num_vehicles"], num_vehicles)
    mean_speeds["tms"] = np.append(mean_speeds["tms"], np.mean(speeds))
    mean_speeds["sms"] = np.append(
        mean_speeds["sms"], num_vehicles / (np.sum(1 / speeds))
    )

mean_speeds_df = pd.DataFrame(mean_speeds)

mean_speeds_df.truncate()
```

```python
import plotly.graph_objects as go

fig = go.Figure(
    layout=go.Layout(
        title="Mean Speeds vs Number of Vehicles",
        xaxis_title="Number of Vehicles",
        yaxis_title="Mean Speed (mi/hr)",
    )
)

fig.add_trace(
    go.Scatter(
        x=mean_speeds["num_vehicles"],
        y=mean_speeds["tms"],
        name="TMS",
    )
)
fig.add_trace(
    go.Scatter(
        x=mean_speeds["num_vehicles"],
        y=mean_speeds["sms"],
        name="SMS",
    )
)

fig.show()
```
